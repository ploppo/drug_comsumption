### Caricamento e Preparazione Base del Dataset

Caricamento del dataset effettuato come nella scorsa esercitazione.

```python
mushrooms = pd.read_csv('Dataset/agaricus-lepiota.data', header=None)

display(mushrooms)
```

#### Assegnazione Nomi a Colonne Dataset

```python
col_names = ['class', 'cap-shape', 'cap-surface', 'cap-color', 'bruises?', 'odor', 'gill-attachment', 'gill-spacing', 
             'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 
             'stalk-color-above-ring', 'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 'ring-type', 
             'spore-print-color', 'population', 'habitat']
rename_dict = {k: col_names[k] for k in range(23)}

mushrooms.rename(columns=rename_dict, inplace=True)

display(mushrooms)
```

#### Conversione dei Dati in "categorical"

```python
mushrooms = mushrooms.astype('category')
```

## Encoding del Dataset

(... saltiamo la parte di esplorazione del dataset, già svolta nell precedente esercitazione...)

**Esercizio:** scrivere nella cella di codice seguente i comandi necessari ad effettuare un one-hot encoding rispetto alle features ed un label encoding rispetto ai target. Il tutto fatto su una _COPIA_ del dataset originale.

```python
# LEGENDA VARIABILI DA USARE:
# mushrooms_work : copia del dataframe mushrooms che verrà trasformato rispetto il nuovo encoding

from sklearn.preprocessing import LabelEncoder
# Copy of the dataframe
mushrooms_work = mushrooms.copy()

# Label Encoding
label_enc = LabelEncoder()
label_enc.fit(mushrooms_work['class'].values)
mushrooms_work['class'] = label_enc.transform(mushrooms_work['class'].values)


# One-hot Encoding
colonne = mushrooms_work.columns.tolist()[1:]
mushrooms_work = pd.get_dummies(mushrooms_work, columns = colonne)

display(mushrooms_work)
```

## Preparazione dei Dati

**Esercizio:** estrarre dal dataframe *mushrooms_work* la matrice $X$ delle features ed il vettore $\boldsymbol{y}$ delle classi (corrispondenti alle righe di $X$). Dopodiché determinare le matrici $X_{\mathrm{train}}, X_{\mathrm{test}}$ ed i vettori $\boldsymbol{y}_{\mathrm{train}}, \boldsymbol{y}_{\mathrm{test}}$ rappresentanti rispettivamente il *training* ed il *test* set. La percentuale di suddivisione del dataser in *training-test* deve essere *50%-50%*; il *random seed* la data di oggi nella forma *yyyymmdd*.

```python
# LEGENDA VARIABILI DA USARE:
# X : matrice delle features di mushrooms_work
# y : vettore delle classi di mushrooms_work
# X_train, X_test: matrici delle features di training e test set, rispettivamente, ottenute da X
# y_train, y_test: vettiri delle classi di training e test set, rispettivamente, ottenuti da y
X = mushrooms_work.iloc[:,1:].values
y = mushrooms_work['class'].values

random_seed = 20210506
test_p = 0.50

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= test_p, random_state = random_seed, shuffle= True)
```

## PCA

**Esercizio:** eseguire la PCA su $X_{\mathrm{train}}$ *senza riduzione della dimensionalità*, per visualizzare l'andamento della varianza spiegata all'aumentare delle PC.

```python
# LEGENDA VARIABILI DA USARE:
# pca_global : oggetto PCA che NON effettua riduzione di dimensionalità

pca_global = PCA()

pca_global.fit(X_train)

plt.figure()

plt.plot(np.insert(np.cumsum(pca_global.explained_variance_ratio_),0,0))
plt.title('Mushrooms')
plt.xticks(ticks = np.arange(1, pca_global.n_features_+1 ),labels=[f'PC{i}' for i in range(1, pca_global.n_features_ + 1)], rotation=45)
plt.xlabel('Principle components')
plt.ylabel('Cumulative explained variance')
plt.grid()
plt.show()
```

**Esercizio:** eseguire la PCA su $X_{\mathrm{train}}$ preservando una percentuale di varianza spiegata pari almeno a *explvar_p=0.90*. Visualizzare il numero di PC selezionate.

```python
# LEGENDA VARIABILI DA USARE:
# pca : oggetto PCA che effettua riduzione di dimensionalità
explvar_p = 0.90
pca = PCA(explvar_p)
pca.fit(X_train)

df_pca = pd.DataFrame({'val':pca.n_components_,'varianza':pca.explained_variance_ratio_.sum()}, index =['n.PC'])
df_pca
#pca.explained_variance_ratio_
```

## FDA

Applichiamo la FDA sia ad $X$ "puro" che alla sua trasformazione di dimensioni ridotte tramite PCA e confrontiamo i risultati.

**Esercizio:** implementa due FDA, una preparata su *(X_train,y_train)* e l'altra sul caso equivalente ma con *X_train* trasformato tramite pca. Trasforma poi *X* rispetto alle due FDA e confronta visivamente i due risultati.

**ATTENZIONE:** Le classi sono 2, quindi la dimensione massima dello spazio per la rappresentazione in dimensione ridotta è... Escogitare, di conseguenza, un modo per agevolare la rappresentazione grafica.

```python
# LEGENDA VARIABILI DA USARE:
# fda : oggetto FDA da applicare ai dati senza PCA
# fda_pca: oggetto FDA da applicare ai dati con PCA
from FisherDA import MultipleFisherDiscriminantAnalysis as MDA
fda = MDA()
fda_pca = MDA()

fda.fit(X_train,y_train)

fda_pca.fit(pca.transform(X_train),y_train)


Zfda = fda.transform(X)
Zfda_pca = fda_pca.transform(pca.transform(X))

fig1, axs1 = plt.subplots(1, 2, figsize=(8, 3))

axs1[0].scatter(Zfda.flatten(), y, c=y, alpha=0.15)

axs1[1].scatter(Zfda_pca.flatten(), y, c=y, alpha=0.15)          
```

## LDA

**Esercizio:** ripetere quanto fatto con la FDA anche per la LDA. Aggiungere però anche i plot ed un calcolo dell'accuratezza della predizione sul training e test set.

```python
# LEGENDA VARIABILI DA USARE:
# lda : oggetto LDA da applicare ai dati senza PCA
# lda_pca: oggetto LDA da applicare ai dati con PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
lda_pca = LDA()

lda.fit(X_train, y_train)
lda_pca.fit(pca.transform(X_train),y_train)

Zlda_X= lda.transform(X)
Zlda_pcaX = lda_pca.transform(pca.transform(X))

Zlda_Xtr= lda.transform(X_train)
Zlda_pcaXtr = lda_pca.transform(pca.transform(X_train))

Zlda_Xte= lda.transform(X_test)
Zlda_pcaXte = lda_pca.transform(pca.transform(X_test))


fig2, axs2 = plt.subplots(1, 2, figsize=(8, 3))
axs2[0].scatter(Zlda_X.flatten(), y, c=y, alpha=0.15)
axs2[1].scatter(Zlda_pcaX.flatten(), y, c=y, alpha=0.15)
axs2[0].grid()
axs2[1].grid()

fig3, axs3 = plt.subplots(1, 2, figsize=(8, 3))
axs3[0].scatter(Zlda_Xtr.flatten(), y_train, c=y_train, alpha=0.15)
axs3[1].scatter(Zlda_pcaXtr.flatten(), y_train, c=y_train, alpha=0.15)
axs3[0].grid()
axs3[1].grid()


fig4, axs4 = plt.subplots(1, 2, figsize=(8, 3))
axs4[0].scatter(Zlda_Xte.flatten(), y_test, c=y_test, alpha=0.15)
axs4[1].scatter(Zlda_pcaXte.flatten(), y_test, c=y_test, alpha=0.15)
axs4[0].grid()
axs4[1].grid()
df_lda = pd.DataFrame({'acc. (no PCA)': [lda.score(X_train, y_train), lda.score(X_test, y_test)],
                       'acc. (PCA)': [lda_pca.score(pca.transform(X_train), y_train), lda_pca.score(pca.transform(X_test), y_test)]}, 
                      index=['training', 'test'])
df_lda
```
